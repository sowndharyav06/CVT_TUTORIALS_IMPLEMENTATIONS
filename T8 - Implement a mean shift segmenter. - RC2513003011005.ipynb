```python
# ============================================
# Mean Shift Segmentation using OpenCV
# ============================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# 1 Upload your iris image (or any image)
uploaded = files.upload()

# 2 Process each uploaded image
for fname in uploaded.keys():
    print(f"\nProcessing image: {fname}")

    # Read image
    img = cv2.imread(fname)
    if img is None:
        print(" Could not load image, skipping.")
        continue

    # Convert to RGB for displaying with matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 3 Apply mean-shift filtering
    # Parameters:
    #   sp: spatial window radius
    #   sr: color window radius
    # Larger values -> stronger smoothing / fewer segments
    mean_shift = cv2.pyrMeanShiftFiltering(img_rgb, sp=20, sr=40)

    # 4 Convert to grayscale and apply thresholding for visualization
    gray = cv2.cvtColor(mean_shift, cv2.COLOR_RGB2GRAY)
    _, segmented = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # 5 Display results
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.imshow(img_rgb)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.imshow(mean_shift)
    plt.title('Mean Shift Filtered')
    plt.axis('off')

    plt.subplot(1,3,3)
    plt.imshow(segmented, cmap='gray')
    plt.title('Segmented Mask')
    plt.axis('off')

    plt.show()

#  End of script
```
