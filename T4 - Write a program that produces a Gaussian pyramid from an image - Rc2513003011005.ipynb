```python
# =====================================================
# Gaussian Pyramid Construction from Uploaded Image
# (Google Colab Ready)
# =====================================================
!pip install -q opencv-python-headless matplotlib numpy

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload your image
print("ðŸ‘‰ Please upload your image (e.g., aeval1.bmp)")
uploaded = files.upload()
if not uploaded:
    raise SystemExit("No image uploaded.")
image_path = next(iter(uploaded.keys()))
print(f"âœ… Uploaded: {image_path}")

# Step 2: Read and convert to RGB
img = cv2.imread(image_path)
if img is None:
    raise ValueError("Error: Could not read image.")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Step 3: Build Gaussian Pyramid
num_levels = 5  # you can change this number
gaussian_pyramid = [img_rgb]
for i in range(1, num_levels):
    img_rgb = cv2.pyrDown(img_rgb)  # smooth + downsample by factor of 2
    gaussian_pyramid.append(img_rgb)

# Step 4: Display the Gaussian Pyramid
plt.figure(figsize=(15, 8))
for i, layer in enumerate(gaussian_pyramid):
    plt.subplot(1, num_levels, i + 1)
    plt.imshow(layer)
    plt.title(f'Level {i}')
    plt.axis('off')
plt.suptitle("Gaussian Pyramid (Multi-scale Image Representation)", fontsize=14)
plt.show()

# Step 5: Save all pyramid levels
for i, layer in enumerate(gaussian_pyramid):
    filename = f"gaussian_level_{i}.png"
    cv2.imwrite(filename, cv2.cvtColor(layer, cv2.COLOR_RGB2BGR))
    print(f"ðŸ’¾ Saved: {filename}")

# Step 6: Offer downloads
for i in range(num_levels):
    files.download(f"gaussian_level_{i}.png")

print("âœ… Gaussian pyramid generation complete.")
```
