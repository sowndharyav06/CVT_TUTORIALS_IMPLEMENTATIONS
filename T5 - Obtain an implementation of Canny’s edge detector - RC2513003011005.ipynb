```python
# ================================
# Human Iris Edge Detection using Canny
# ================================

# 1 Install and import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import files

# 2 Upload image(s) from your local system
# You can upload a single image or multiple iris images
uploaded = files.upload()

# 3 Process each uploaded image
for fname in uploaded.keys():
    print(f"\nProcessing image: {fname}")

    # Read image using OpenCV
    img = cv2.imread(fname)
    if img is None:
        print(" Could not load image; skipping.")
        continue

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Optional: Denoise to improve edge clarity
    blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)

    # 4 Apply Canny Edge Detection
    # You can tune threshold1 and threshold2 for sensitivity
    edges = cv2.Canny(blurred, threshold1=50, threshold2=150)

    # 5 Display results
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.imshow(gray, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')

    plt.subplot(1,3,3)
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edge Detection')
    plt.axis('off')

    plt.show()

#  End of script
```
